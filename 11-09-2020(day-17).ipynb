```python
# matplotlib
- raw data converts into graphical representation.
- data visulization
-# installtion:
- pip install matplotlib
# for using:
import matplotlib.pyplot as plt

- format strings:
    [color,linestyle,marker]
- https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html
```

```python
# basic linear plot
import matplotlib.pyplot as plt
x=[25,26,27,28,29]
y1=[28000,29000,30000,31000,32000]
plt.plot(x,y1,label="Y1_axis",marker="d",linestyle="--")
y2=[29000,32000,45000,50000,55000]
plt.plot(x,y2,label="Y2_axis",color="y",marker="^",linestyle="-.")
plt.tight_layout()
plt.grid(True)
plt.legend()
plt.xlabel('ages')
plt.ylabel('salary')
plt.title('basic linear plot')
#print(plt.style.available)
plt.style.use('fivethirtyeight')
#plt.savefig('plot.png')
plt.show()
```

```python
# subplot
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(1,10,100)
plt.figure(1)

plt.subplot(2,2,1)
plt.title('linear')
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title('cubic')
plt.plot(x,x**3)

plt.figure(2)

plt.subplot(2,2,1)
plt.title('log')
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title('exponential')
plt.plot(x,x**2)

plt.show()
```
